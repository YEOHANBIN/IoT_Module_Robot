```python
from pingpongthread import PingPongThread
import time
```

```python
# 1개 로봇을 연결하는 인스턴스 생성.
PingPong = PingPongThread(number=1, group_id=57)                                         
```

```python
# 로봇 제어 스레드 시작.
PingPong.start()
# 모든 로봇이 연결될 때 까지 기다림.
PingPong.wait_until_full_connect() 
```

```python
# 핑퐁 센서 수신 설정(주기 0.1로 설정)
PingPong.receive_sensor_data(1, method="periodic", period=0.1)
```

```python
# 버튼을 누르지 않으면 계속 반복하기
while not PingPong.get_current_button(1) == 1:
    #자이로 센서값 수신하여 gyro_value에 저장하기
    gyro_value = PingPong.get_current_gyro(1)
    #현재 자이로 roll 센서값 출력
    print("roll: ",gyro_value[0])
    #현재 자이로 pitch 센서값 출력
    print("pitch: ",gyro_value[1])
    #현재 자이로 yaw 센서값 출력
    print("yaw: ",gyro_value[2])
    time.sleep(0.5) # 0.5초 기다리기    
```

```python
#2. roll값을 gyro_value[0]을 통해 네모방향
#또는 별 방향으로 기울어짐 감지하고
#해당 방향으로 회전시키기
while not PingPong.get_current_button(1) == 1: 
    gyro_value = PingPong.get_current_gyro(1)    
    # 자이로 센서 roll값 출력
    print(gyro_value[0])
    # 네모 방향으로 빠르게 움직이면 "네모 방향으로 회전!"를 출력
    # 별 방향으로 빠르게 움직이면 "별 방향으로 회전!" 출력
    if gyro_value[0] < -50 :
        print("네모 방향으로 회전!")
        PingPong.run_motor(1,30)
    elif gyro_value[0] > 50 :
        print("별 방향으로 회전!")
        PingPong.run_motor(1,-30)
    # 0.1초 기다리기
    time.sleep(0.1)
#버튼을 클릭하면 모터도 멈추기
PingPong.run_motor(1,0)
```

```python
#센서값 수신 종료
PingPong.stop_sensor_data(1)
```

```python
#로봇 제어 스레드 종료
PingPong.end()
```
